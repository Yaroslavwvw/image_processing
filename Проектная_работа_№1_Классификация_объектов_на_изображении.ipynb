```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Импортируем нужную функцию для Colab

# Загрузка изображения
img = cv2.imread('coins2.jpg')  # Замените 'coins.jpg' на ваш файл изображения

# Преобразование в градации серого
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Размытие изображения для уменьшения шума
gray_blurred = cv2.medianBlur(gray, 5)

# Применение преобразования Хафа для обнаружения кругов
dp = 1.2                # Параметр разрешения аккумуляторного массива
minDist = 100           # Минимальное расстояние между центрами обнаруженных окружностей
param1 = 50             # Порог для метода Кэнни
param2 = 150            # Порог для функции HoughCircles (чем меньше, тем больше ложных кругов)
minRadius = 30          # Минимальный радиус круга (можно уменьшить)
maxRadius = 250         # Максимальный радиус круга (можно увеличить)

# Обнаружение кругов
circles = cv2.HoughCircles(
    gray_blurred, cv2.HOUGH_GRADIENT, dp, minDist,
    param1=param1, param2=param2,
    minRadius=minRadius, maxRadius=maxRadius
)

# Проверка, найдены ли круги
if circles is not None:
    circles = np.uint16(np.around(circles))

    # Получение всех радиусов
    radii = [circle[2] for circle in circles[0, :]]

    # Находим минимальный и максимальный радиус
    min_radius = min(radii)
    max_radius = max(radii)

    # Распределение монет по номиналам по заданным правилам
    coin_values = []
    for radius in radii:
        # Если радиус отличается от минимального не более чем на 9%
        if radius <= min_radius * 1.09:
            coin_values.append(1)  # 1 рубль

        # Если радиус отличается от максимального не более чем на 7%
        elif radius >= max_radius * 0.93:
            coin_values.append(5)  # 5 рублей

        else:
            coin_values.append(2)  # 2 рубля

    # Выводим список найденных монет
    print("Общая сумма:", sum(coin_values))

    # Отображение найденных кругов и номиналов с радиусами
    img_circles = img.copy()
    for i, circle in enumerate(circles[0, :]):
        center = (circle[0], circle[1])
        radius = circle[2]
        cv2.circle(img_circles, center, radius, (0, 255, 0), 2)  # Рисование контура круга
        cv2.circle(img_circles, center, 2, (0, 0, 255), 3)  # Рисование центра круга
        # Добавление текста с номиналом и радиусом
        cv2.putText(img_circles, f"{coin_values[i]} RUB, R={radius}",
                    (center[0] - 30, center[1] + 10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.5, (255, 0, 0), 1, cv2.LINE_AA)

else:
    print("Круги не найдены")

# Отображение результатов через imshow в Colab
cv2_imshow(img_circles)
```

```python
import cv2
import numpy as np

def process_frame(frame, previous_circles, frame_count, update_interval=10):
    """Обрабатывает кадр для определения монет и их номиналов."""
    # Используем предыдущее значение кругов, если не обновляем
    if frame_count % update_interval != 0 and previous_circles is not None:
        return annotate_frame(frame, previous_circles)

    # Преобразование в градации серого
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Размытие для уменьшения шума
    gray_blurred = cv2.medianBlur(gray, 5)

    # Параметры для Хафа
    dp = 1.2
    minDist = 50
    param1 = 100
    param2 = 40
    minRadius = 20
    maxRadius = 100

    # Обнаружение кругов
    circles = cv2.HoughCircles(
        gray_blurred, cv2.HOUGH_GRADIENT, dp, minDist,
        param1=param1, param2=param2,
        minRadius=minRadius, maxRadius=maxRadius
    )
    if circles is not None:
        # Округление координат
        circles = np.uint16(np.around(circles))

    return circles

def annotate_frame(frame, circles):
    """Добавляет аннотации (номиналы, сумма) на кадр."""
    if circles is None:
        return frame  # Если кругов нет, просто возвращаем кадр

    radii = [circle[2] for circle in circles[0, :]]

    # Определение самого большого и самого маленького радиуса
    min_radius = min(radii)
    max_radius = max(radii)

    coin_values = []
    filtered_circles = []
    for circle in circles[0, :]:
        radius = circle[2]
        # Условие фильтрации кругов по радиусу
        if min_radius * 0.9 <= radius <= max_radius * 1.1:
            filtered_circles.append(circle)
            if radius <= min_radius * 1.09:
                coin_values.append(1)  # 1 рубль
            elif radius >= max_radius * 0.95:
                coin_values.append(5)  # 5 рублей
            else:
                coin_values.append(2)  # 2 рубля

    # Рисование кругов и номиналов
    for i, circle in enumerate(filtered_circles):
        center = (circle[0], circle[1])  # Координаты центра
        radius = circle[2]  # Радиус
        cv2.circle(frame, center, radius, (0, 255, 0), 2)  # Контур круга
        cv2.circle(frame, center, 2, (0, 0, 255), 3)  # Центр круга
        # Нанесение текста с номиналом
        cv2.putText(frame, f"{coin_values[i]} RUB",
                    (center[0] - 30, center[1] + 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1, cv2.LINE_AA)

    # Отображение общей суммы
    total = sum(coin_values)
    cv2.putText(frame, f"Total: {total} RUB",
                (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
                1, (0, 255, 255), 2, cv2.LINE_AA)

    return frame

def main():
    """Главная функция для работы с веб-камерой и записи видео."""
    cap = cv2.VideoCapture(0)  # 0 — индекс камеры

    if not cap.isOpened():
        print("Не удалось открыть веб-камеру.")
        return

    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS)) or 30

    out = cv2.VideoWriter(
        'coin_detection_filtered.avi',
        cv2.VideoWriter_fourcc(*'XVID'),
        fps,
        (frame_width, frame_height)
    )

    print("Нажмите 'q' для завершения записи.")

    previous_circles = None
    frame_count = 0
    update_interval = 10  # Частота обновления кругов (каждые 10 кадров)

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Не удалось получить кадр с веб-камеры.")
            break

        frame_count += 1
        circles = process_frame(frame, previous_circles, frame_count, update_interval)

        # Если круги обновились, сохраняем их
        if frame_count % update_interval == 0:
            previous_circles = circles

        # Аннотируем кадр
        processed_frame = annotate_frame(frame, previous_circles)
        out.write(processed_frame)

        cv2.imshow("Монеты: распознавание и расчет суммы", processed_frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    out.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
